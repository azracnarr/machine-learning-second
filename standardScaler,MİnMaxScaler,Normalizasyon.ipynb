```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("plane.csv")
```

```python
df.info()
```

```python
df=df[["Rcmnd cruise Knots","Stall Knots dirty","Fuel gal/lbs","Eng out rate of climb","Takeoff over 50ft","Price"]]
```

```python
df.head(3)
```

```python
df=df.dropna() #none değerleri atar
```

```python
y=df["Price"]
x=df.drop("Price",axis=1)
```

```python
x_train,y_train,x_test,y_test=train_test_split(x,y,random_state=23,train_size=0.76)
```

```python
ss=StandardScaler()
modelss=ss.fit_transform(x_train,y_train)
```

```python
modelss=pd.DataFrame(modelss)
modelss.head(3)
```

```python
modelss[4].mean()
```

```python
modelss[4].std()
```

```python
mm=MinMaxScaler()
modelmm=mm.fit_transform(x_train,y_train)
```

```python
modelmm=pd.DataFrame(modelmm)
```

```python
modelmm.head(3)
```

```python
modelmm[3].max()
```

```python
modelmm[3].min()
```
