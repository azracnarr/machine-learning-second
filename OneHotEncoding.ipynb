```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv("churn.csv")
```

```python
df.info()
```

```python
df.head(3)
```

Verdiğimiz öğretmeler sonucu müşterinin çıkıp çıkmadığını kontrol edeceğiz.


```python
df=df.drop(["RowNumber","CustomerId", "Surname"],axis=1)
```

```python
df.head(3)
```

geography and gender are abject biz bunları int yapmaya çalışacaz dummy kullanarak önceki hali bu şekildeydi ama şimdi onehotencoder kullanarak yapacağız
df=pd.get_dummies(df,columns=["Geography","Gender"],drop_first=True)

```python
ohe=OneHotEncoder()
xd=ohe.fit_transform(df[["Geography","Gender"]]).toarray()
```

Datamızdaki dummies edilmesi gereken yerler az olduğundan colonda çok fazla artış olmadı ama fazla dummies edilen verilerde hotEncoding kullanılır.

```python
ohe.get_feature_names_out()
```

yeni bir dataFrame oluşturup birleştirme işlemi yapalım.

```python
y=pd.DataFrame(xd)
y.columns=ohe.get_feature_names_out()
```

```python
y.head(3)
```

```python
df.head(3)
```

dfmizdeki geography and gender kısmını silip yeni dataFramemizi oraya eklememiz gerekiyor.

```python
df=df.drop(columns=["Geography","Gender"])
```

```python
df.head(3)
```

```python
df[y.columns]=y
```

```python
df
```

```python
y=df["Exited"]
x=df.drop("Exited",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=23,train_size=0.75)
```

```python
rf=RandomForestClassifier()
model=rf.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree=DecisionTreeClassifier()
model=tree.fit(x_train,y_train)
model.score(x_test,y_test)
```
