```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns   #hazır veriseti içerir veri görselleştirmede kullanılır
from sklearn.model_selection import train_test_split   #x_train x_test y_train y_test metotları için
```

```python
df=sns.load_dataset("diamonds")
```

```python
df.head()
```

```python
df.columns=["karat","kesim","renk","kalite","derinlik","masa","fiyat","x","y","z"] #isim değişikliği atama yapma dikkat et
```

```python
df.head()
```

```python
df=pd.get_dummies(df,columns=["kesim","renk","kalite"],drop_first=True, dtype=int)
```

```python
df.head(10)
```

```python
y=df["fiyat"]
x=df.drop("fiyat",axis=1)
```

x_train ve y_train bizim modelde kullandığımız verilerin öğrenme oranı içi
x_test ve y_test modeldeki verilerimizi test edeceğimiz gerçek hayataki gibi verri girişleri rolünü oynuyor.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=23,train_size=0.74)
```

```python
lm=LinearRegression()
model=lm.fit(x_train,y_train)
lm.score(x_test,y_test)  #model score için test girişlerine bakılır
```

Ama eğitim için olan trainx ve y'yi score alırsak çıkan sonuçlar arasındaki fark (x ve y Test ile aralarındaki fark) bize model hakkında overfitting mi underfitting mi yoksa balancedfittin mi bilgi verir. overfitting ezber yapmak aradaki fark çok olur.Under fittin aradaki fark az olur ama tam istenilen gibi olmaz.İyileştirmeler yaparak underfitting'i Balancedfittinge yaklaştırmaya çalışırız.
